## Question 1: [Codeforces 4A. Watermelon](https://codeforces.com/problemset/problem/4/A)
#### Lance B.
This question requires a tiny bit of math, but it is definitely manageable. They key to this solution is that an even number (boy number one's part) plus an even number (boy number two's part) must equal another even number. This way, we can simply output "YES" if the input is a positive even number greater than $2$ ($2$ cannot be created via two positive even numbers), and "NO" otherwise. We check if a number is divisible by $2$ using the modulo operator. Here are the solution codes:

### Python:
```
w = int(input())
if w % 2 == 0 and w > 2:
    print("YES")
else:
    print("NO")
```
### C++:
```
#include <iostream>
using namespace std;

int main(){
    int w; cin >> w;
    if(w % 2 == 0 && w > 2) cout << "YES\n";
    else cout << "NO\n";
}
```

## Question 2: [Leetcode 1. Two Sum](https://leetcode.com/problems/two-sum/)
#### Lance B.
For this question, for each element of the list/array, iterate through everything after it (convince yourself that this does not repeat any pairings of indices, but goes through every one) and check if their sum is equal to the target. Here are the solution codes:

### Python:
```
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]
```
### C++:
```
class Solution {
public:
    vector<int> twoSum(vector<int>& nums, int target) {
        for(int i = 0; i < nums.size(); i++){
            for(int j = i + 1; j < nums.size(); j++){
                if(nums[i] + nums[j] == target) return vector<int> {i, j};
            }
        }
        return {}; // Dummy
    }
};
```

## Question 3: [Leetcode 9. Palindrome Number](https://leetcode.com/problems/palindrome-number/)
#### Lance B.
One easy way would be to convert the input into a string and use the built-in reverse functions to see if the string and its reverse are the same.

### Python:
```
class Solution:
    def isPalindrome(self, x: int) -> bool:
        x = str(x)
        reversed = x[::-1]
        return x == reversed
```
### C++:
```
class Solution {
public:
    bool isPalindrome(int x) {
        string dummy = to_string(x);
        string reversed = dummy;
        reverse(reversed.begin(), reversed.end());
        return dummy == reversed;
    }
};
```


But this seems a bit cheap. Instead, we could also use a *start* and *end* pointer to check that each place is equal to itself on the opposite side.

### Python:
```
class Solution:
    def isPalindrome(self, x: int) -> bool:
        x = str(x)
        start = 0
        end = len(x) - 1
        while(start < end):
            if x[start] != x[end]:
                return False
            start += 1
            end -= 1
        return True
```
### C++:
```
class Solution {
public:
    bool isPalindrome(int x) {
        string dummy = to_string(x);
        int start = 0, end = dummy.size() - 1;
        while(start < end){
            if(dummy[start] != dummy[end]) return false;
            start++; end--;
        }
        return true;
    }
};
```


However, there's also another, more challenging, solution without converting the paramter into a string. This would be to continuously get the digit in $x$'s ones place, add it to another variable, say $y$, multiply $y$ by $10$, then divide $x$ by $10$. This way, we can compare $y$ to our original parameter, and we should get the same answer. This is essentially "reversing" the number using math. Make sure to have another variable to keep track of what your original paramter was.

### Python:
```
class Solution:
    def isPalindrome(self, x: int) -> bool:
        compare = x
        y = 0
        while(x > 0):
            y += x % 10
            x = int(x / 10)
            y *= 10
        # Note that y was multiplied by an unnecessary 10 at the end
        y /= 10
        return y == compare
```
## C++:
```
class Solution {
public:
    bool isPalindrome(int x) {
        int compare = x;
        long long y = 0; // Note that y may overflow if it were an int, so make it a long long
        while(x > 0){
            y += x % 10;
            x /= 10;
            y *= 10;
        }
        // Note that y was multiplied by an unnecessary 10 at the end
        y /= 10;
        return y == compare;
    }
};
```

## Question 4: [Codeforces 50A. Domino Piling](https://codeforces.com/problemset/problem/50/A)
#### Lance B.
For this question, let's consider four cases. If the $m$ and $n$ are both even, we can pack as many dominos as we want without any going out-of-bounds. If the $m$ is even and $n$ is odd, we can still pack as many as we want without any going out-of-bounds. An odd $m$ and even $n$ uses the same logic. As of right now, our formula is simply $\frac{m \cdot n}{2}$. However, if both $m$ and $n$ are odd, we cannot include that half domino that would be sticking out; we'd have to subtract $\frac{1}{2}$ from our current formula. Luckily, we can simply take the floor of $\frac{m \cdot n}{2}$, and we won't have to do anything extra for the odd $m$ by odd $n$ case. Therefore, our final formula is $[{\frac{m \cdot n}{2}}]$, where $[x]$ is the greatest integer $\le x$.

### Python:
```
integers = input().split() # Split the input into a list

m = int(integers[0]) # Get the values and convert to integers
n = int(integers[1])

ans = int(m * n / 2) # Note that int() takes the floor of a value

print(ans)
```

### C++:
```
#include <iostream>
using namespace std;

int main(){
    int m, n; cin >> m >> n;
    int ans = m * n / 2; // Note that C++ ints automatically take the floor
    cout << ans << '\n';
}
```

## Question 5: [USACO 2016 Bronze: Square Pasture](http://usaco.org/index.php?page=viewproblem2&cpid=663)
#### Lance B.
Believe or not, this question is quite simple! All we have to do is keep track of the biggest and smallest $x$ and $y$ values. The biggest $x$ and $y$ values would form the top right corner of our rectangle, and the smallest $x$ and $y$ values would form the bottom left corner of our rectangle. We can then use the differences between the $x$ and $y$ values to calculate the length and width of the rectangle. We must then take the bigger of these two (to form a square), and square it to get our answer.

### Python:
```
fin = open("square.in", 'r')

bigx, bigy, smallx, smally = -1, -1, 11, 11 # Start off with extremes

for i in range(2): # Because there are two lines
    line = fin.readline().split()
    for j in range(2): # For the two coordinates in a line
        currx = int(line[j * 2])
        curry = int(line[j * 2 + 1])
        bigx = max(bigx, currx)
        bigy = max(bigy, curry)
        smallx = min(smallx, currx)
        smally = min(smally, curry)

length = bigx - smallx
width = bigy - smally

side = max(length, width)

fout = open("square.out", 'w')
fout.write(str(side ** 2) + '\n')
```

### C++:
```
#include <fstream>
#include <iostream>
#include <cmath>
using namespace std;

int main(){
    ifstream fin ("square.in"); ofstream fout ("square.out");
    int bigx = -1, bigy = -1, smallx = 11, smally = 11; // Start off with extremes
    for(int i = 0; i < 4; i++){ // Because there are four coordinates
        int currx, curry; fin >> currx >> curry;
        bigx = max(bigx, currx);
        bigy= max(bigy, curry);
        smallx = min(smallx, currx);
        smally = min(smally, curry);
    }
    int length = bigx - smallx;
    int width = bigy - smally;
    int side = max(length, width);
    fout << side * side << '\n';
}
```
