## Question 1: [Leetcode 136. Single Number](https://leetcode.com/problems/single-number/)
#### Lance B.
First, create a dictionary/map to keep track of the occurrences for each number. Then, iterate through the keys of your dictionary/map and return the key that has value one.
### Python:
```
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        occurrences = dict()
        for i in nums:
            if i in occurrences: # Python dictionaries are a bit annoying; you have to initialize each key
                occurrences[i] += 1
            else:
                occurrences[i] = 1
        for key, value in occurrences.items():
            if value == 1:
                return key
```
### C++:
```
class Solution {
public:
    int singleNumber(vector<int>& nums) {
        unordered_map<int, int> occurrences; // Use unordered maps if order of your keys doesn't matter; runs in O(1)
        for(int i : nums){
            occurrences[i]++;
        }
        for(auto it : occurrences){
            if(it.second == 1) return it.first;
        }
        return -1; // Dummy
    }
};
```

## Question 2: [Leetcode 242. Valid Anagram](https://leetcode.com/problems/valid-anagram/)
#### Lance B.
For this question, keep track of the number of times each character in $s$ occurs. After that, iterate through $t$ and make sure that the number of occurrences of each character matches the number of occurrences in $s$.
### Python:
```
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if len(s) != len(t):
            return False # If their lengths are different, they're not anagrams
        occurrences = dict()
        for i in s:
            if i in occurrences:
                occurrences[i] += 1
            else:
                occurrences[i] = 1
        for i in t:
            if i in occurrences:
                occurrences[i] -= 1
            else:
                return False
        for key, value in occurrences.items():
            if value != 0:
                return False
        return True
```
### C++:
```
class Solution {
public:
    bool isAnagram(string s, string t) {
        if(s.size() != t.size()) return false;
        unordered_map<char, int> occurrences;
        for(char c : s){
            occurrences[c]++;
        }
        for(char c : t){
            if(occurrences[c] == 0) return false;
            occurrences[c]--;
        }
        for(auto it : occurrences){
            if(it.second != 0) return false;
        }
        return true;
    }
};
```

## Question 3: [USACO 2017 Bronze: Don't Be Last!](http://www.usaco.org/index.php?page=viewproblem2&cpid=687)
Given the constraints of this problem, we don't have to worry much about time. First, we'll want to initialize all of the cows with production of $0$. After that, go through the input and increase the cows' milk produced accordingly. We'll then want to get the smallest value, and use this to get the second-smallest value. After this, we see which cows produced the second-smallest value; if there's only one, output its name, output "Tie" otherwise.
#### Lance B.
### Python:
```
fin = open("notlast.in", 'r')
fout = open("notlast.out", 'w')

# Prepare the dictionary
cow = ["Bessie", "Elsie", "Daisy", "Gertie", "Annabelle", "Maggie", "Henrietta"]
cows = dict()
for x in cow:
  cows[x] = 0

# Get the inputs
N = int(fin.readline().strip())
for x in range(N):
  line = fin.readline().split()
  name = line[0]
  amount = int(line[1])
  cows[name] += amount

# Get the smallest produced
smallest = 1000000000
for k, v in cows.items():
  smallest = min(smallest, v)

# Get the second-smallest produced; the "desired" number
secondSmallest = 1000000000
for k, v in cows.items():
  if v > smallest:
    secondSmallest = min(secondSmallest, v)

# Keep track of the desired cows
desired = []
for k, v in cows.items():
  if v == secondSmallest:
    desired.append(k)

# Output
if len(desired) == 1:
  fout.write(desired[0] + '\n')
else:
  fout.write("Tie\n")
```
### C++:
```
#include <iostream>
#include <fstream>
#include <string>
#include <vector>
#include <unordered_map>
using namespace std;

int main(){
    // Intialize map
    unordered_map<string, int> cows;
    string names[7] = {"Bessie", "Elsie", "Daisy", "Gertie", "Annabelle", "Maggie", "Henrietta"};
    for(int i = 0; i < 7; i++){
        cows[names[i]] = 0;
    }

    // Get inputs
    ifstream fin ("notlast.in");
    int N; fin >> N;
    for(int i = 0; i < N; i++){
        string cow; int production;
        fin >> cow >> production;
        cows[cow] += production;
    }

    // Get the smallest
    int smallest = 1e9;
    for(auto it : cows){
        smallest = min(smallest, it.second);
    }

    // Get the second smallest
    int secondSmallest = 1e9;
    for(auto it : cows){
        if(it.second > smallest){
            secondSmallest = min(secondSmallest, it.second);
        }
    }

    // Keep track of desired cows
    vector<string> desired;
    for(auto it : cows){
        if(it.second == secondSmallest){
            desired.push_back(it.first);
        }
    }

    // Output
    ofstream fout ("notlast.out");
    if(desired.size() == 1){
        fout << desired[0] << '\n';
    }
    else{
        fout << "Tie\n";
    }
}
```
